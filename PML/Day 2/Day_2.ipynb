List : Hetrogeneous, ordered, mutable and duplicates allowed

```python
ordered_items =["Pizza","Burger", "Pasta","Pizza"]
ordered_items
```

```python
ordered_items.append("Burger")
ordered_items.extend(["Coke","ICe cream"])
```

```python
ordered_items
```

```python
ordered_items.remove("Burger") # Remove by name
ordered_items
```

```python
ordered_items.pop() # remove last element
ordered_items
```

```python
ordered_items.pop(2)
ordered_items
```

```python
ordered_items[:2] #Slicing
```

```python
for item in ordered_items:
    print(item)
```

```python
upper_case = [item.upper() for item in ordered_items]
upper_case
```

Dictionary : Key-value, hetrogeneous, no order, mutable and duplicate key not allowed.

```python
customer = {"name" : "Raj",
            "age": 22,
            "city":"Mumbai"}
customer
```

```python
customer["name"]
```

```python
customer.get("age")
```

```python
customer["phone"] = 999999991
customer["age"] = 23
```

```python
customer
```

```python
del customer["age"]
customer
```

```python
for key, value in customer.items():
    print(key,value)
```

Sets : unordered, Unique, not mutable

```python
cuisines = {"Indian","Chinese","Italian","Indian"}
cuisines
```

```python
cuisines.add("Mexican")
cuisines
```

```python
cuisines.discard("Chinese")
```

```python
cuisines_set1 = {"Indian","Chinese"}
cuisines_set2 = {"Indian", "Mexican","Italian"}
```

```python
cuisines_set1.union(cuisines_set2) # Union
```

```python
cuisines_set1.intersection(cuisines_set2)  #intersection
```

```python
cuisines_set1.difference(cuisines_set2) # difference
```

Tuple: Ordered, Immutable

```python
location =(19.0700,72.8777)
lat, lon = location # unpacking
print(lat, lon)
```

```python
t1 = (1,) # single value declaration in tuple
```

```python
df['Grade'] = ['A','B','A']
df
```

```python
df = df.drop('Grade', axis=1)
```

```python
df
```

```python
df.rename(columns={'Name':'Title'}, inplace=False)
```

```python
df['Marks'] = df['Marks'].apply(lambda x: x+5)
```
